```python
#import basic libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import os
```

```python
data = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/00471/Data_for_UCI_named.csv")
#data = pd.read_csv("C:\\Users\\Abdulrazaq\\Documents\\py learn\\HAMOYE\\Hamoye stage C\\Data_for_UCI_named.csv")
```

```python
data
```

```python
X = data.drop(columns =["stabf"])
y = pd.DataFrame()
y["stabf"] = data["stabf"]
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test, y_train,y_test = train_test_split(X,y, test_size = 0.3,
                                                 random_state = 42)
```

```python
X_train.shape,X_test.shape, y_train.shape,y_test.shape
```

```python
#Question 14:accuracy of testset with random forest
from sklearn.ensemble import RandomForestClassifier
rand = RandomForestClassifier().fit(X_train,y_train)
rand_pred =rand.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
acc =(accuracy_score(rand_pred,y_test))
round(acc,4)
```

```python
#Question 15
import xgboost
from xgboost import XGBClassifier
xg = XGBClassifier().fit(X_train,y_train)
xg_pre=xg.predict(X_test)
acc_xg =(accuracy_score(xg_pre,y_test))
round(acc_xg,4)
```
